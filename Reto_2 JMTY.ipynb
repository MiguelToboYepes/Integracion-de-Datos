```python
# Variables de Entrada: Distribución de las Pérdidas
# Variables de Salida: Limite Inferior, Limite Superior, Número de Datos
# Marca de Clase del Histograma de Frecuencias.
def Histx(LDA): #Nombre de su método

  NC=10  #Número de Clusters

  plt.figure()
  sns.histplot(LDA,kde=True,bins=NC,color='blue')
  plt.title('Distribución LDA')
  plt.show()

  #Obtenemos la información de la Distribución
  counts, bin_edges=np.histogram(LDA,bins=NC)
  LI=bin_edges[:-1];LS=bin_edges[1:]
  mc=(LI+LS)/2
  freq=counts/np.sum(counts)
  print(LI)
  print(LS)
  print(counts)
  print(mc)
  print(freq)

  return LI,LS,counts,mc,freq
```

# 3. Se cargan los datos de trabajo

```python
import pandas as pd
nxl=('/content/drive/MyDrive/BI /6. SolicitantesCrédito(USD) (1).xlsx')

XDB=pd.read_excel(nxl,sheet_name=1)
display(XDB)

#Se calcula la distribucion
LDA=XDB.iloc[:,25]*XDB.iloc[:,33]*(1+XDB.iloc[:,35])
XDB['LDA'] = LDA
display(LDA)

LI,LS,counts,mc,freq=Histx(LDA)
df=pd.DataFrame({'LI':LI,'LS':LS,'counts':counts,'mc':mc,'freq':freq})
display(df)
df.to_excel('Histograma_LDA.xlsx')
```

# 4. Muestreo aleatorio: Utilizando el metodo de montecarlo

```python
acum=np.cumsum(freq)
Confiabilidad=9999+1 #Confiabilidad del 99,99%
Confiabilidad=int(Confiabilidad)
print(Confiabilidad)
print(acum)
LDAm=np.zeros((Confiabilidad,1))
```

```python
for k in range(Confiabilidad):
  al=np.random.uniform(0,1)
  inter=np.where(al<acum)[0][0]
  LDAm[k,]=np.random.uniform(LI[inter],LS[inter])
```

```python
plt.figure()
sns.histplot(LDAm,kde=True,bins=10,color='blue')
sns.histplot(LDA,kde=True,bins=10,color='pink')
plt.title('Distribución LDAm (Muestreada)')
plt.show()
```

5. Prueba de muestreo

```python
from scipy.stats import skew,kurtosis,ttest_ind
print('Informacion LDA original')
print('Media:', np.mean(LDA))
print('Varianza:', np.var(LDA))
print('Desviacion:', np.std(LDA))
print('Ceficiente de asimetria:', skew(LDA))
print('Curtosis:', kurtosis(LDA))

print('informacion LDA muestreada')
print('Media:', np.mean(LDAm))
print('Varianza:', np.var(LDAm))
print('Desviacion:', np.std(LDAm))
print('Coeficiente de asimetria:', skew(LDAm))
print('Curtosis:', kurtosis(LDAm))
```

```python
#Las diferencias entre las metricas son:
dif_u = ((np.mean(LDA) - np.mean(LDAm)) / np.mean(LDA)) * 100
print('La diferencia porcentual de la media es:',dif_u)
dif_varianza = ((np.var(LDA) - np.var(LDAm)) / np.var(LDA)) * 100
print('La diferencia porcentual de la varianza es:',dif_varianza)
dif_desviacion = ((np.std(LDA) - np.std(LDAm)) / np.std(LDA)) * 100
print('La diferencia porcentual de la desviacion es:',dif_desviacion)
dif_asimetria = ((skew(LDA) - skew(LDAm)) / skew(LDA)) * 100
print('La diferencia porcentual de la asimetria es:',dif_asimetria)
dif_curtosis = ((kurtosis(LDA) - kurtosis(LDAm)) / kurtosis(LDA)) * 100
print('La diferencia porcentual de la curtosis es:',dif_curtosis)
```

Tras realizar el muestreo Monte Carlo con 10 000 iteraciones, comparamos los estadísticos obtenidos de la muestra (LDAm) con los valores de la población original (LDA). A continuación, el análisis actualizado:

+ Media: La media de la muestra (55.69) es muy cercana a la media original (55.37). La diferencia porcentual es de 0.38 %, lo que confirma una excelente aproximación de la tendencia central.

+ Varianza: La muestra presenta una varianza de 800.70 frente a 756.09 en los datos originales. La diferencia porcentual es de -4.80 %, lo que sugiere una leve sobreestimación de la dispersión, pero en términos absolutos sigue siendo aceptable.

+ Desviación Estándar: Calculada en 28.30 para la muestra contra 27.50 del original. La diferencia porcentual es de -2.37 %, una pequeña variación que mantiene la consistencia general de la distribución.

+ Asimetría (Skewness): El valor muestral es de 0.39 frente a 0.47 en la población. La diferencia porcentual es 6.97 %, lo que indica que la muestra conserva el sesgo positivo hacia la derecha, aunque con una forma ligeramente más atenuada.

+ Curtosis: La muestra muestra una curtosis de 0.07, mientras que el original era de 0.22, con una diferencia porcentual de -11.64 %. Esto implica que la muestra tiene una distribución más achatada y colas menos pronunciadas, lo que podría subestimar eventos extremos —un factor importante en análisis de riesgo.

 La muestra generada mediante Monte Carlo logra reflejar muy bien la media de la distribución original, y con discrepancias pequeñas en los otros resultados. Las diferencias en varianza, desviación estándar, asimetría y curtosis son pequeñas. La curtosis es importante , ya que su impacto potencial sobre la estimación del riesgo extremo. Para afinar la precisión en las colas de la distribución, podriamos considerar aumentar el número de simulaciones.
